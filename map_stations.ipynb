```python
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.cm import ScalarMappable
from mpl_toolkits.basemap import Basemap
# from matplotlib.animation import FuncAnimation
```

```python
stations = pd.read_csv("stations.csv")
stations.info()
```

```python
locations = stations['Location'].to_list()
latitudes = stations['Latitude'].to_list()
longitudes = stations['Longitude'].to_list()
```

```python
# Create a figure and draw a basemap
fig, ax = plt.subplots(figsize=(10,8), dpi=500)

map = Basemap(resolution='h',
            projection='lcc',
            lat_0=16.33, lon_0=121.15, # Center
            llcrnrlat=12.06, # The lower left corner geographical latitude
            llcrnrlon=118.87, # The lower left corner geographical longitude
            urcrnrlat=18.93, # 	The upper right corner geographical latitude
            urcrnrlon=124.91 # The upper right corner geographical longitude
            )

map.drawcoastlines()
map.drawmapboundary(zorder=0)
map.fillcontinents(color='#ffffff',zorder=1)
map.drawcountries(linewidth=1.5)

x, y = map(longitudes, latitudes)
map.scatter(x, y, marker='D', color='red', zorder=2)

plt.title("Northern and Central Luzon Weather Stations")
plt.show()
```
